```python
import cv2

def resize_image(input_path, output_path, size=(512, 512)):
    # Read the image
    img = cv2.imread(input_path)

    # Resize the image
    resized_img = cv2.resize(img, size)

    # Save the resized image
    cv2.imwrite(output_path, resized_img)

# Example usage
input_image_path = '/content/drive/MyDrive/ENPM809K-Final Project/segmenter/random images'
output_image_path = '/content/drive/MyDrive/ENPM809K-Final Project/segmenter/random images'

resize_image(input_image_path, output_image_path)
```
